```python
import cv2
import os
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import numpy as np
```

```python
jpeg_dataset_folder = r'C:\Users\Anwender\Downloads\jpeg_pics'
no_traffic_sign_recognition = r'C:\Users\Anwender\Downloads\no_traffic_sign_recognition'

#set bbox saver
bbox_gt_path = r'C:\Users\Anwender\Downloads\no_traffic_sign_recognition\gt.txt'
bboxes = []

withmuchrred = 0
```

```python
def get_red_value(pixels, threshhold):
    #pixels should be np.float32

    mean_red = np.mean(pixels[:, :, 0])
    red_percentage = (mean_red / 255) * 100

    if red_percentage < threshhold:
        return True
    else:
        return False
```

```python
def split_and_find_bboxs(image, filename):
    #image should be cv2 image
    print(filename)

    height, width, _ = image.shape

    #get size for quatering the image
    quadrant_height = height // 2
    quadrant_width = width // 2

    #get random bboxes in quadrants
    for y in range(2):
        for x in range(2):
            top = y * quadrant_height 
            bottom = top + quadrant_height
            left = x * quadrant_width
            right = left + quadrant_width

            #set coordinats for bboxs
            random_size = np.random.randint(20, 120)
            random_top = np.random.randint(top, bottom - random_size)
            random_bottom = random_top + random_size
            random_left = np.random.randint(left, right - random_size)
            random_right = random_left + random_size


            #presave bbox for prosessing 
            bbox_data = image[random_top:random_bottom, random_left:random_right, :].astype(np.float32)

            while_break = 0

            global withmuchrred


            #evaluate bbox and if necessary replace
            while get_red_value(bbox_data, 90) == False:
                #reset coordinats for bboxs
                random_size = np.random.randint(20, 120)
                random_top = np.random.randint(top, bottom - random_size)
                random_bottom = random_top + random_size
                random_left = np.random.randint(left, right - random_size)
                random_right = random_left + random_size

                #represave bbox for prosessing 
                bbox_data = image[random_top:random_bottom, random_left:random_right, :].astype(np.float32)

                while_break += 1

                if(while_break == 100):
                    withmuchrred += 1
                    print('no bbox with little red found')
                    break

            while_break = 0
            
            #save the bbox
            bboxes.append(str(random_right) + ';' + str(y) + str(x) + ';' + str(random_top) + ';' + str(random_bottom)  + ';' + str(random_left)  + ';' + str(random_right))
            image_path = os.path.join(no_traffic_sign_recognition, filename + str(y) + str(x) + '.jpg')
            cv2.imwrite(image_path, bbox_data) 
            print('saved: ' + image_path)
```

```python
def go_through_images():
    for root,_,files in os.walk(jpeg_dataset_folder):
    
        for file in files:
            if file.endswith('.jpg'):
                image = cv2.imread(os.path.join(root, file))
                split_and_find_bboxs(image, str(file).split('.')[0])
```

```python
go_through_images()
print(withmuchrred)
with open(bbox_gt_path, 'w') as file:
    for bbox in bboxes:
        file.write(bbox + "\n")
```
