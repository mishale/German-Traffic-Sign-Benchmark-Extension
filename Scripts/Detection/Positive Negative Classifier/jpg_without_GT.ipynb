```python
import cv2
import os
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
```

```python
#set the Dataset link
dataset_folder = r'C:\Users\Anwender\Downloads\FullIJCNN2013\FullIJCNN2013'
groundtruth_bbox = r'C:\Users\Anwender\Downloads\FullIJCNN2013\FullIJCNN2013\gt.txt'
jpeg_dataset_folder = r'C:\Users\Anwender\Downloads\jpeg_pics'
```

```python
def pic_with_groundtruth(groundtruth_bbox):
    withGT = []
    with open(groundtruth_bbox, 'r') as file:
        lines = file.readlines()
        for line in lines:
            withGT.append(line.split(';')[0])
            #print(line.split(';')[0])
    print(len(withGT))

    return withGT
```

```python
def save_as_JPEG(withGT):

    for root, _, files in os.walk(dataset_folder):
        _.clear()
        for file in files:
            toSave = False
            if any(str(file) in picName for picName in withGT):
                toSave = False
            else:
                toSave = True
            
            if toSave:
                jpg_image = os.path.join(jpeg_dataset_folder, str(file).split('.')[0] + '.jpg')
                file = Image.open(os.path.join(dataset_folder, str(file)))
                file.save(jpg_image, "JPEG")
                print('Saved as ' + jpg_image)
```

```python
withGT = pic_with_groundtruth(groundtruth_bbox)
save_as_JPEG(withGT)
```
