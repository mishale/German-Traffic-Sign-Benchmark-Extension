```python
import csv
import os
```

```python
def ppm_to_csv_train(directory, label_file, output_file):
    csv_data = []
    
    with open(label_file, 'r') as file:
        labels = file.readlines()
        labels = [label.strip() for label in labels]
        print(len(labels))
        labels = [label.split(';')[-1] for label in labels]


    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".ppm"):
                image_path = os.path.join(root, file)
                label = labels[int(file.split(".")[0])]
                csv_data.append([image_path, label])
    
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Path', 'Label'])
        writer.writerows(csv_data)
```

```python
def  ppm_to_csv_test(directory, output_file):
    csv_data = []

    for x in range(43):
        label = str(x)
        if x < 9:
            label = '0' + label
        directory_path = os.path.join(directory, label)
        for root, _, files in os.walk(directory_path):
            for file in files:
                if file.endswith(".ppm"):
                    image_path = os.path.join(root, file)
                    csv_data.append([image_path, label])

    
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Path', 'Label'])
        writer.writerows(csv_data)
```

```python
def ppm_to_csv_fullscreen(directory, boundingbox_file, test_output_file, train_output_file):
    csv_dataTrain = []
    csv_dataTest = []

    with open(boundingbox_file, 'r') as file:
        boundingboxs = file.readlines()
        boundingboxs = [boundingbox.strip() for boundingbox in boundingboxs]
        ToT = 0 #Test or Training
        boundingboxsTest = []
        boundingboxsTrain = []
        current_pic = boundingboxs[0].split(';')[0]
        for boundingbox in boundingboxs:
            boundingboxs_pic = boundingbox.split(';')

            if boundingboxs_pic[0] != current_pic:
                ToT += 1
                current_pic = boundingboxs_pic[0]
                while ToT != int(current_pic.split('.')[0], 10):
                    ToT += 1
            elif boundingboxs_pic[0] == current_pic:
                current_pic = boundingboxs_pic[0]

            if ToT < 600:
                boundingboxsTrain.append(str(ToT) + ':' + boundingboxs_pic[1] + ';' + boundingboxs_pic[2] + ';' + boundingboxs_pic[3] + ';' +boundingboxs_pic[4])
            else:
                boundingboxsTest.append(str(ToT) + ':' + boundingboxs_pic[1] + ';' + boundingboxs_pic[2] + ';' + boundingboxs_pic[3] + ';' + boundingboxs_pic[4])
                 
          

    for root, _, files in os.walk(directory):
        ToT = 0
        current_pic = ToT
        testCheck = 0
        _.clear()
        for file in files:
            if file.endswith(".ppm"):
                if ToT == 600:
                    testCheck = 600
                    current_pic = 0
                if  ToT < 600:
                    image_pathTrain = os.path.join(root, file)
                    while current_pic < len(boundingboxsTrain) and ToT == int(boundingboxsTrain[current_pic].split(':')[0]):
                        boundingboxTrain = boundingboxsTrain[current_pic].split(':')[1]
                        csv_dataTrain.append([image_pathTrain, boundingboxTrain])
                        current_pic += 1
                    ToT += 1
                else:    
                    image_pathTest = os.path.join(root, file)
                    while current_pic < len(boundingboxsTest) and testCheck == int(boundingboxsTest[current_pic].split(':')[0]):
                        boundingboxTest = boundingboxsTest[current_pic].split(':')[1]
                        csv_dataTest.append([image_pathTest, boundingboxTest])
                        current_pic += 1
                    ToT += 1
                    testCheck += 1
                
                
    
    with open(train_output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Path', 'Bonding Boxs'])
        writer.writerows(csv_dataTrain)

    with open(test_output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Path', 'Bonding Boxs'])
        writer.writerows(csv_dataTest)
```

```python
def read_ppm(file_path):
    with open(file_path, 'rb') as ppm_file:
        ppm_file.readline()  # Skip the PPM file header
        dimensions = ppm_file.readline().decode().rstrip().split()
        width, height = int(dimensions[0]), int(dimensions[1])
        max_val = int(ppm_file.readline().rstrip())

        byte_data = ppm_file.read()

    return byte_data, width, height, max_val
```
